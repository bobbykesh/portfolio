```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Customer Loyalty History.csv')
```

```python
data.columns = data.columns.str.replace(' ','_').str.lower()
data.head(10)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
during_campaign = data.query(('(enrollment_year == 2018) & (enrollment_month in [2,3,4]) '))['enrollment_year']. value_counts().sum()
```

```python
before_campaign = data.query(('(enrollment_year == 2017) & (enrollment_month in [2,3,4]) '))['enrollment_year'].value_counts().sum()
```

```python
during_campaign 
```

```python
label = ['During Campaign','Same Months Before Campaign']
label_data = [during_campaign,before_campaign]
```

```python
plt.figure(figsize  =(10,10))
plt.bar(label,label_data,color = 'skyblue', edgecolor = 'black', width = 0.5)
plt.title('Distribution  of Customers Enrollment During Campaign Period And Before')
plt.ylabel('Total Enrollment')
plt.grid(True,ls = ':',c = 'black')
plt.show()
```

```python
cancellations = data.query('cancellation_year ==2018 & cancellation_month in [2,3,4] ')['enrollment_year'].value_counts().sum()
cancellations
```

```python
new_enrollment = data.query('(enrollment_year == 2018) & (enrollment_month in [2,3,4]) '). value_counts().sum()
new_enrollment
```

```python
net_label = ['Cancellations','Enrollments']
net_data = [cancellations,new_enrollment]
```

```python
plt.figure(figsize  =(10,10))
plt.bar(net_label,net_data,color = 'skyblue',  edgecolor = 'black',width =0.5)
plt.title('Analysis During Campaign')
plt.ylabel('Count')
plt.grid(True,ls = ':',c = 'black')
plt.show()
```

```python
gender_during_campaign = data.query('enrollment_year ==2018 & enrollment_month in [2,3,4]'). groupby('gender')['enrollment_year'].value_counts()
```

```python
plt.figure(figsize  = (10,10))
gender_during_campaign.plot(kind = 'barh', color= 'skyblue',  edgecolor  = 'black')
plt.title('Distribution Of Enrollments Based On Gender During Campaign')
plt.ylabel('')
plt.xlabel('Count')
plt.grid(True,ls = ':',c = 'black')
plt.show()
```

```python
province_during_campaign = data.query('enrollment_year == 2018 & enrollment_month in [2,3,4] ').groupby('province')['enrollment_year'].value_counts()
```

```python
plt.figure(figsize = (10,10))
province_during_campaign.plot(kind = 'barh' ,color ='skyblue', edgecolor = 'black')
```
